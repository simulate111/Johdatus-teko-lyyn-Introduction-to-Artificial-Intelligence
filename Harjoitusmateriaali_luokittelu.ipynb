### Luokittelijan suorituskyvyn arvioiminen

Luokittelussa luokittelijan suorituskykyä arvioidaan testidatan sekaannusmatriisista (engl. confusion matrix) lasketuilla mitta-arvoilla. Sekaannusmatriisi kertoo, kuinka moni näytteistä luokitellaan oikein ja kuinka moni väärin sekä mihin luokkiin väärin luokitellut näytteet oikeasti kuuluvat. Tässä harjoitusmateriaalissa sekaannusmatriisissa oikeat luokat on sijoitettu riveille ja luokittelijan ennustamat luokat sarakkeille, mutta sijoittelu voidaan tehdä toisinpäinkin. Mikäli sekaannusmatriisin tietyllä rivillä olevalla luokalla on lukuja eri sarakkeilla kuin sen oma rivinumero, on luokittelija luokitellut näytteitä vääriin luokkiin. Kuvassa 1 on esimerkki eräästä sekaannusmatriisista, johon kuuluu 9 luokkaa. Paras tulos luokittelijalle saavutetaan, kun sen sekaannusmatriisin päädiagonaalilla ovat kaikki luokittelijalla ennustetut testidatan luokat.

Mittasuureena luokittelutarkkuus ei kuitenkaan yksistään anna riittävän hyvää kuvaa luokittelijan suorituskyvystä varsinkaan silloin, kun kyseessä on epäbalansoitu data. Epäbalansoidussa datassa eri luokissa on selvästi eri määrä näytteitä. Kun todellisessa tilanteessa käsiteltävänä on epäbalansoitu data, ensimmäinen vaihe on yrittää saattaa se balanssiin yli- tai alinäytteistyksellä. Aina se ei kuitenkaan ole mahdollista. Epäbalansoidun datan luokittelussa käytetään luokittelutarkkuuden sijaan mittoja sensitiivisyys/saanti/herkkyys (engl. sensitivity, recall, true positive rate), spesifisyys (engl. specificity, true negative rate), positiivinen ennustearvo/tarkkuus (engl. positive predictive value, precision) ja negatiivinen ennustearvo (engl. negative predictive value) kuvaamaan luokittelijan suorituskykyä. Käydään seuraavaksi esimerkin avulla läpi, miten eri suorituskyvyn mitta-arvot eroavat toisistaan.

100 ihmisen kylästä 20 sairastuu harvinaiseen tautiin, jonka oireet alkavat näkyä vasta viikon kuluttua tartunnan saamisesta. Kylän lääkärin tehtävä on selvittää taudin havaitsemista varten suunnitellulla laitteella tartunnan saaneet henkilöt, jotta hoito voidaan aloittaa mahdollisimman aikaisessa vaiheessa. Laite arvioi, että 30 ihmistä 100:sta on sairastunut tautiin. Viikon kuluttua kuitenkin selviää, että ennustetuista 30 sairaasta ihmisestä todellisuudessa 15 oli taudinkantajia (TP) ja 15 olikin terveitä (FP). Koska sairastuneita oli yhteensä 20, niin 5 ihmistä tunnistettiin terveiksi, vaikka he olivat oikeasti sairaita (FN). Kylän ihmisistä 65 tunnistettiin oikein terveiksi (TN). Nyt voidaan laskea laitteen suorituskykyä kuvaavat mitta-arvot käyttämällä yhtälöitä (1), (2), (3), (4) ja (5). Kuvassa 3 on havainnollistettu tekstin esimerkki kuvasarjalla. Kuvassa 4 tekstin esimerkille on muodostettu sekaannusmatriisi.

* **TP**, oikea hälytys (engl. true positives, hit)
* **FP**, väärä hälytys (engl. false positives, false alarm)
* **FN**, väärä hylkäys (engl. false negatives, miss)
* **TN**, oikea hylkäys (engl. true negatives, correct rejection)

Sensitiivisyys/saanti/herkkyys:
<br><br>
\begin{equation}
\frac{TP}{P} = \frac{TP}{TP+FN} = \frac{15}{15+5} = 75 \% \tag{1}
\end{equation}

Spesifisyys:
<br><br>
\begin{equation}
\frac{TN}{N} = \frac{TN}{TN+FP} = \frac{65}{65+15} = 81,25 \% \tag{2}
\end{equation}

Positiivinen ennustearvo/tarkkuus:
<br><br>
\begin{equation}
\frac{TP}{TP+FP} = \frac{15}{15+15} = 50 \% \tag{3}
\end{equation}

Negatiivinen ennustearvo:
<br><br>
\begin{equation}
\frac{TN}{TN+FN} = \frac{65}{65+5} = 92,9 \% \tag{4}
\end{equation}

Luokittelutarkkuus/virheettömyys:
<br><br>
\begin{equation}
\frac{TP+TN}{P+N} = \frac{TP+TN}{TP+FN+TN+FP} = \frac{15+65}{15+5+65+15} = 80 \% \tag{5}
\end{equation}
<br><br>

Binääristä kahden luokan luokitteluongelmaa voidaan tarkastella todennäköisyystiheysjakaumina. Tällöin luokkien näytteet toisistaan erottava luokkaraja määritellään numeroarvoisena kynnysarvona. Kynnysarvoa säätämällä voidaan pienentää esimerkiksi väärien hylkäysten määrää (FN) samalla kasvattaen väärien hälytysten määrää (FP). Eri kynnysarvoista laskettuna voidaan muodostaa ROC-käyrä (engl. receiver operating charasterictic curve), jossa pystyakselilla on sensitiivisyys ja vaaka-akselilla 1-spesifisyys (engl. FPR) [1]. Kuvassa 5 on kahden luokan todennäköisyystiheysjakaumat ja niitä vastaava ROC-käyrä. Esimerkiksi vakavaa sairautta tunnistaessa sensitiivisyyden arvo, joka kertoo kuinka moni kaikista sairastuneista henkilöistä (TP+FN) tunnistettiin sairaiksi (TP), on oltava mahdollisimman iso. Tämä toteutuu valitsemalla kuvan 5 ROC-käyrältä piste, jossa sensitiivisyys saa mahdollisimman suuren arvon, mikä näkyy samalla 1-spesifisyyden (FPR) arvon kasvamisena (eli spesifisyyden arvon pienenemisenä). Sensitiivisyyden arvon kasvattaminen ROC-käyrällä vastaa kynnysarvon siirtämistä vasemmalle todennäköisyystiheysjakaumien kuvaajassa.

Paras tulos luokittelijalle saavutetaan, kun ROC-käyrä kulkee vasemman yläkulman pisteen (0,1) kautta. Jos ROC-käyrä kulkee vasemman alakulman pisteen (0,0) ja oikean yläkulman pisteen (1,1) kautta lineaarisesti, on luokittelija yhtä hyvä kuin jokaiselle testinäytteelle arvottaisiin luokka sattumanvaraisesti. ROC-käyrän tulos voidaan muuttaa numeroarvoiseksi laskemalla ROC AUC-arvo (engl. area under the receiver operating charasteristic), joka kertoo kuinka suuri alue jää ROC-käyrän alapuolelle, kun molempien akselien arvot vaihtelevat välillä 0 ja 1. Täydellisen luokittelijan ROC AUC-arvo on 100 % ja sattumanvaraisen luokittelijan ROC AUC-arvo on 50 %. Kuvassa 6 on täydellisen luokittelijan ROC-käyrä vihreällä, sattumanvaraisen luokittelijan ROC-käyrä sinisellä ja niitä vastaavat ROC AUC-arvot.

Toinen tapa arvioida luokittelijan suorituskykyä eri kynnysarvoilla on positiivinen ennustearvo - sensitiivisyys käyrä (engl. precision-recall curve). Siinä vaaka-akselilla on sensitiivisyys ja pystyakselilla positiivinen ennustearvo. Positiivinen ennustearvo - sensitiivisyys käyrää tulee käyttää erityisesti silloin, kun käsitellään epäbalansoitua dataa. ROC-käyrä antaa epäbalansoidulla datalla testattaessa luokittelijan suorituskyvystä liian optimistisen kuvan. Positiivinen ennustearvo - sensitiivisyys käyrään ei datan epäbalanssilla ole erityisen suurta vääristävää vaikutusta. Täydellisen luokittelijan positiivinen ennustearvo - sensitiivisyys käyrä kulkee oikean yläkulman pisteen (1,1) kautta. Myös positiivinen ennustearvo - sensitiivisyys käyrän tulos voidaan muuttaa numeroarvoiseksi PR AUC-arvoksi (engl. area under the precision recall curve) laskemalla käyrän alapuolinen alue. Kuvassa 7 on täydellisen luokittelijan positiivinen ennustearvo - sensitiivisyys käyrä, jonka PR AUC-arvo on 100 %.
